```python
df1 = pd.value_counts(df['INCIDENT_TYPE_DESC'])/sum(pd.value_counts(df['INCIDENT_TYPE_DESC']))*100
data2 = {'Incident_type': df1.keys(),'Ratio': df1[df1.keys()]}
df2 = pd.DataFrame(data2)
df2 = df2[df2['Ratio']>0.7]
r = 100-df2['Ratio'].sum()
labels = df2['Incident_type'].unique()
```

```python
pie, ax = plt.subplots(figsize=[20,20])
labels = df2['Incident_type'].unique()
plt.pie(x=df2['Ratio'], autopct="%.1f%%", labels=labels, pctdistance=.6,startangle=70)
plt.title("Types of incidents", fontsize=15);
plt.show()
```
